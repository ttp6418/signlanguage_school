```python
import pandas as pd
import numpy as np
import requests
import urllib.request
import io
import time

import warnings
warnings.filterwarnings('ignore')

from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
```

<h2>crawling 함수 정의</h2>

```python
video_list = pd.DataFrame(index=range(0), columns={'no', 'url', 'kor', 'part'})

def video_url_crawling(num):
    options = webdriver.ChromeOptions()
    options.add_argument('headless')
    options.add_argument("disable-gpu") 
    options.add_argument("disable-infobars")
    options.add_argument("--disable-extensions")
    driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)
    driver.implicitly_wait(5)
    driver.get(f'http://sldict.korean.go.kr/front/sign/signContentsView.do?origin_no={num}')
    driver.implicitly_wait(5)

    html = driver.page_source
    soup = BeautifulSoup(html,'html.parser')
    
    sub_cat = soup.select_one('#depthmenu > ul > li:nth-child(5) > a > span').text
    value = soup.select_one('#signViewForm > dl > dd:nth-child(2)').text.strip().split(',')
    part = soup.select_one('#signViewForm > dl > dd:nth-child(3)').text
    
    
    if sub_cat == '국어 교과 용어':
        video_url = soup.find(type="video/mp4").get("src")
        print(video_url)
        print(f"{num}번 영상 : {value} : {part}")
        
        global video_list
        video_list = video_list.append(pd.DataFrame({'no':num, 'url':video_url, 'kor':value, 'part': part}))
        video_list.reset_index(drop=True, inplace=True)
        
    return video_list
```

<h2><p>반복문을 활용한 크롤링</p></h2>

```python
# 전문용어 수어 中 국어 교과 교육 중 조사랑 어미 등... (19500 ~ 20900)

for num in range(1, 27000):
    try:
        video_url_crawling(num)
    except:
        print(num, "통과")
        continue
    print("===========================================================================")
```

```python
import moviepy
import time
from tqdm import tqdm
from moviepy.editor import VideoFileClip

## 518번 계속 403에러 
## 지속적인 URL 요청으로 인해 크롤링 제한인줄,,
## 크롤링시 URL 자체가 비어있던 문제 

vd = df["url"]
sec = []

for i in tqdm(range(len(vd))):
    try:
        clip = VideoFileClip(vd[i])
        sec.append(clip.duration)
        time.sleep(1)
    except:
        print("error_num:",i)
        sec.append("NaN")
        
```
